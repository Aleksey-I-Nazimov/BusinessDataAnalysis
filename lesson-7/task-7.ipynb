Предсказание успешности студентов: https://archive.ics.uci.edu/dataset/697/predict+students+dropout+and+academic+success

```python
import pandas as pd
import numpy as np
```

Чтение и подготовка данных

```python
data = pd.read_csv("data.csv")

# All columns are represented as a string. It has to be corrected:--------------
split_titles=[]
for title in data.columns[0].split(";"):
    split_titles.append(title)

# Each ROW is a string. It has to be splitted:----------------------------------
def split_row (row) :
    not_split_values = row[0]
    row_values = []
    for row_value in not_split_values.split(";"):
        row_values.append(row_value)
    return row_values;

split_data = data.apply(func=split_row, axis=1)

data_array = None
for index,item in split_data.items():
    if data_array is None:
        data_array = np.array(item)
    else :
        data_array = np.vstack([data_array,np.array(item)])

df = pd.DataFrame(data=data_array,columns=split_titles)

df.head()
        
```

```python
df.info()
```

```python
TARGET = 'Target'
df[TARGET].value_counts()
```

```python
new_df = df.copy()

# Changing target to binary:-------------------------------------------------
new_df.loc[new_df[TARGET] == 'Graduate', 'Target'] = "1"
new_df.loc[new_df[TARGET] == 'Dropout', 'Target'] = "0"

# Excluding 'Enrolled':------------------------------------------------------
new_df = new_df[(new_df[TARGET]=='1') | (new_df[TARGET]=='0')]
new_df[TARGET].value_counts()

# Changing types:------------------------------------------------------------
for column in new_df.columns:
    if column != TARGET:
        new_df[column] = new_df[column].astype(float)
    else :
        new_df[column] = new_df[column].astype(int)
#new_df.info()
```

```python
new_df.nunique()
```

Нулей не имеем. Значения расставлены по колонкам. Типы колонок приведены к числовым. Бинаризация target произведена<br>


```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(new_df.drop([TARGET],axis=1),
                                                    new_df[TARGET],random_state=2,stratify=new_df[TARGET],test_size=0.51)

print (x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler

from xgboost import XGBClassifier


# Setting pipe for transformers:-----------------------------------------------------
class NumberSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # The key can refer to name only. In this case only one is selected
        return X[[self.key]]

transformers = list();
for column in x_train.columns:
    pipe = Pipeline([
        ('selector', NumberSelector(key=column)),
        ('scaler', StandardScaler())
    ])
    transformers.append((column,pipe))
    
    
# Getting XGBoost:-------------------------------------------------------------------
boosClassifier = XGBClassifier(n_estimators=5, max_depth=8, learning_rate=1, objective='binary:logistic',random_state=42)   
    
feature_union = FeatureUnion(transformers);
pipeline = Pipeline([
    ("features",feature_union),
    ("classifier", boosClassifier)
])


# Launching the FIT:-----------------------------------------------------------------
pipeline.fit(x_train,y_train)
    
```

```python
from sklearn.metrics import precision_recall_curve

y_pred = pipeline.predict_proba(x_test)[:,1]

precision,recall,thresholds = precision_recall_curve(y_test,y_pred)

fscore = (2*precision*recall)/(precision+recall)

ix = np.argmax(fscore)
print (thresholds[ix],fscore[ix],precision[ix],recall[ix])
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

labels=["Dropout","Graduated"]

confusion = confusion_matrix(y_test, y_pred>=thresholds[ix])

disp = ConfusionMatrixDisplay(confusion_matrix=confusion,display_labels=labels)

disp.plot();
```

```python
# The BASE student conversion:------------------------------------------
TN = confusion[0][0]
FN = confusion[1][0]
TP = confusion[1][1]
FP = confusion[0][1]

c0 = (TP+FN)/np.sum(confusion)
print ("The basic studen conversion: {}".format(c0))
```
