```python
# All columns are represented as a string. It has to be corrected:-----------------------------------
split_titles=[]
for title in data.columns[0].split(";"):
    split_titles.append(title)
split_titles
```

```python
# Each ROW is a string. It has to be splitted:----------------------------------------
def split_row (row) :
    not_split_values = row[0]
    row_values = []
    for row_value in not_split_values.split(";"):
        row_values.append(row_value)
    return row_values;

split_data = data.apply(func=split_row, axis=1)
data_array = None
for index,item in split_data.items():
    if data_array is None:
        data_array = np.array(item)
    else :
        data_array = np.vstack([data_array,np.array(item)])
data_array
    
```

```python
df = pd.DataFrame(data=data_array,columns=split_titles)
df
```

```python
df.info()
```

```python
TARGET = 'Target'
df[TARGET].value_counts()
```

```python
new_df = df.copy()

# Changing target to binary:-------------------------------------------------
new_df.loc[new_df[TARGET] == 'Graduate', 'Target'] = "1"
new_df.loc[new_df[TARGET] == 'Dropout', 'Target'] = "0"

# Excluding 'Enrolled':------------------------------------------------------
new_df = new_df[(new_df[TARGET]=='1') | (new_df[TARGET]=='0')]
new_df[TARGET].value_counts()

```

```python
new_df.nunique()
```

```python
for column in new_df.columns:
    if column != TARGET:
        new_df[column] = new_df[column].astype(float)
    else :
        new_df[column] = new_df[column].astype(int)
new_df.info()
```

```python
from sklearn.model_selection import train_test_split



x_train, x_test, y_train, y_test = train_test_split(new_df.drop([TARGET],axis=1),
                                                    new_df[TARGET],random_state=2,stratify=new_df[TARGET],test_size=0.51)

x_train, x_valid, y_train, y_valid = train_test_split(x_train,y_train,random_state=2,stratify=y_train,test_size=0.5)

print (x_train.shape,x_valid.shape,x_test.shape,y_train.shape,y_valid.shape,y_test.shape)

```

```python
from statsmodels.stats.weightstats import ztest

columns_to_be_dropped=[]     # this columns will be dropped
threshold=0.1;               # 2 x 0.05 for pvalue

for column in new_df.drop([TARGET],axis=1).columns:
    z_score, z_pvalue = ztest(x_train[column],x_valid[column])
    print ("  --> {} z pvalue = {}".format(column,z_pvalue))
    if z_pvalue<threshold:
        columns_to_be_dropped.append(column)
        print ("      > It will be dropped")
    
for dropped_column in columns_to_be_dropped:
    x_train=x_train.drop([dropped_column],axis=1)
    x_valid=x_valid.drop([dropped_column],axis=1)
    x_test=x_test.drop([dropped_column],axis=1)
    
print ("  --> Dropped columns {}".format(columns_to_be_dropped))
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier


class NumberSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # The key can refer to name only. In this case only one is selected
        return X[[self.key]]

transformers = list();
for column in x_train.columns:
    pipe = Pipeline([
        ('selector', NumberSelector(key=column)),
        ('scaler', StandardScaler())
    ])
    transformers.append((column,pipe))
    
feature_union = FeatureUnion(transformers);
pipeline = Pipeline([
    ("features",feature_union),
    ("classifier", RandomForestClassifier(random_state=42))
])

pipeline.fit(x_train,y_train)
    
```

```python
from sklearn.metrics import precision_recall_curve

y_pred = pipeline.predict_proba(x_valid)[:,1]

precision,recall,thresholds = precision_recall_curve(y_valid,y_pred)

fscore = (2*precision*recall)/(precision+recall)

ix = np.argmax(fscore)
print (thresholds[ix],fscore[ix],precision[ix],recall[ix])
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

labels=["Graduated","Dropout"]

confusion = confusion_matrix(y_valid, y_pred>=thresholds[ix])

disp = ConfusionMatrixDisplay(confusion_matrix=confusion,display_labels=labels)

disp.plot();
```

```python
# The BASE student conversion:------------------------------------------
TN = confusion[0][0]
FN = confusion[1][0]
TP = confusion[1][1]
FP = confusion[0][1]

c0 = (TP+FN)/np.sum(confusion)
print ("The basic studen conversion: {}".format(c0))
```

```python
x_a,x_b,y_a,y_b = train_test_split(x_test,y_test, test_size=0.5, random_state=42, stratify=y_test)
print (x_train.shape,x_valid.shape,x_test.shape,y_train.shape,y_valid.shape,y_test.shape)
```

```python
y_pred_b = pipeline.predict_proba(x_b)[:,1]
confusion = confusion_matrix(y_b, y_pred_b>=thresholds[ix])
TN = confusion[0][0]
FN = confusion[1][0]
TP = confusion[1][1]
FP = confusion[0][1]

disp = ConfusionMatrixDisplay(confusion_matrix=confusion,display_labels=labels)
disp.plot();
```

```python
# Conversion for A group without ML:-------------------------------------
result_a = (np.sum(y_a))/y_a.shape[0]
print("  --> A group conversion (GRADUATED): {}".format(result_a))

# Conversion for the group B:--------------------------------------------
user_reaction_coefficient = 0.5
result_b = (FN + (1-user_reaction_coefficient) * TP) / np.sum(confusion)
print ("  --> B group conversion (PREDICTED GRADUATED): {}".format(result_b))
```
